**Am going to use column transformer for my full pipeline**

```
from sklearn.compose import ColumnTransformer
num = list(num_attr)
cat = list(train_cat)

full_pipeline = ColumnTransformer([
                                   ("num_transform",num_pipeline,num),
                                   ("cat_transform",cat_pipeline,cat)
])
```
